```python
## SOLVING THE PROBLEM USING PYTHON
words = []
with open("shakespeare.txt", "r") as f:
    for line in f:
        words.extend(line.split())
from collections import Counter
counts = Counter(words)
K = 50
topK = counts.most_common(K)
print(topK)
```
